```python
#import libraries
import pandas 
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
#read dataset
df = pandas.read_csv('play1.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
#assign data to X(features) and y(label) variables
features = ['OUTLOOK','TEMPERATURE','HUMIDITY','WIND']
x = df[features]
y = df['PLAY']
```

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(x,y)
```

```python
fig = plt.figure(figsize=(8,8))
_ = tree.plot_tree(dtree, 
                   feature_names=features,  
                   class_names=['0','1'],
                   filled=True)
```

```python
#predict new values
print(dtree.predict([[1,0,0,0]]))
```

```python
print(dtree.predict([[0,2,1,0]]))
```

```python
score =dtree.score(x,y)
```
